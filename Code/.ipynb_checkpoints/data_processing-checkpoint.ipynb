```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/Tous votes Constituante - Sheet1.csv")
df.iloc[0]
```

```python
#filter the data set and only take 1ère lecture
filtered_df = df.drop(['startTime', 'endTime', 'filename','Unnamed: 0', 'affair','meaningYesText', 'meaningNoText', 'note', 'warning', 'linkVideo', 'attachment', 'DE_meaningYesText', 'DE_meaningNoText', 'DE_note', 'DE_affair', 'DE_section', 'DE_warning', 'DE_attachment'], axis=1)
filtered_df = filtered_df[filtered_df['section'].str.contains("1ère lecture")]
filtered_df.iloc[0]
```

```python
filtered_df['art_id'] = filtered_df['label'].apply(lambda x: x.rsplit(' ')[0])
filtered_df.head()
```

```python
filtered_df.size
```

```python
#Check this part !!!!!!!!!
filtered_df = filtered_df[filtered_df['art_id'].str[0].str.isnumeric()]
```

```python
filtered_df.size
```

```python
filtered_df = filtered_df.drop_duplicates(subset=['art_id'], keep='first')
```

```python
filtered_df.head()
```

```python
ids_vote = filtered_df['affairVoteId'].to_csv('votation_ids.csv')
```

```python
def political_orientation(x):
    if x == 'PDC' or x == 'SVPO' or x == 'CSPO' or x =='CVPO':
        return 'CD'
    elif x == 'VLR' or x == 'UDC':
        return 'D'
    else:
        return 'G'
def vote_transformation(x):
    if x == 'Oui/Ja':
        return 'oui'
    elif x == 'Non/Nein':
        return 'non'
    elif x == 'Abst./Enth.':
        return 'abst'
    else: 
        return 'aucun'
```

```python
df1 = pd.DataFrame(['D', 'CD', 'G'])
df2 = pd.DataFrame(['oui', 'non', 'aucun', 'abst'])
orientation_df = df1.merge(df2, how='cross')
orientation_df.columns=['group', 'vote']
orientation_df.head()
```

```python
ids_vote = filtered_df['affairVoteId']
full_votes = pd.DataFrame()
for ids in ids_vote:
    temp_df = pd.read_csv("../Data/{}.csv".format(ids))
    #for group, vote in orientation_df.enumerate():
        
    full_votes = pd.concat([full_votes, temp_df])
temp_df = full_votes
temp_df['group'] = temp_df['group'].apply(lambda x: political_orientation(x))
temp_df['vote'] = temp_df['vote'].apply(lambda x: vote_transformation(x))
print(temp_df)
```

```python
ori_df = pd.DataFrame()
columns = []
for group, vote in zip(orientation_df.group, orientation_df.vote):
    ori_df[("{}, {}".format(group, vote))]=""
ori_df['affairVoteId'] = ""
ori_df.head()
```

```python
def add_elems(x):
    if x[1] == 'D':
        if x[2] == 'oui':
            return 'D, oui'
        elif x[2] == 'non':
            return 'D, non'
        elif x[2] == 'aucun':
            return 'D, aucun'
        else :
            return 'D, abst'
    elif x[1] == 'CD':
        if x[2] == 'oui':
            return 'CD, oui'
        elif x[2] == 'non':
            return 'CD, non'
        elif x[2] == 'aucun':
            return 'CD, aucun'
        else :
            return 'CD, abst'
    else: 
        if x[2] == 'oui':
            return 'G, oui'
        elif x[2] == 'non':
            return 'G, non'
        elif x[2] == 'aucun':
            return 'G, aucun'
        else :
            return 'G, abst'
```

```python
counts_df = temp_df.groupby(['affairVoteId','group', 'vote']).size()
index = -1
art_id = 0
for ids, val in zip(counts_df.keys(), counts_df):
    temp_art_id = ids[0]
    if(art_id != temp_art_id):
        #print(ids)
        index += 1
        ori_df.at[index,'affairVoteId'] = ids[0]
        art_id = temp_art_id

    if ids[1] =='CD':
        if ids[2] == 'oui':
            ori_df.at[index,"{}".format(add_elems(ids))] = 0.5*val/54
        elif ids[2] == 'non':
            #print(-(0.3*val/54))
            ori_df.at[index,"{}".format(add_elems(ids))] = -(0.5*val/54)
        else:
            ori_df.at[index,"{}".format(add_elems(ids))] = 0
    elif ids[1] =='G':
        if ids[2] == 'oui':
            ori_df.loc[index,"{}".format(add_elems(ids))] = -(val/41)
        elif ids[2] == 'non':
            ori_df.loc[index,"{}".format(add_elems(ids))] = val/41
    else :
        if ids[2] =='oui':
            ori_df.at[index,"{}".format(add_elems(ids))] = val*0.5/35
        elif ids[2] == 'non':
            ori_df.at[index,"{}".format(add_elems(ids))] = -(0.5*val/35)
ori_df = ori_df.fillna(0)
ori_df.head(15)
```

```python
#We divide by two to normalize since we are taking into account the no votes
final_ori_df = pd.DataFrame(columns=['affairVoteId', 'orientation'])
final_ori_df['affairVoteId'] = ori_df['affairVoteId']
final_ori_df['orientation'] = (ori_df['D, oui'] + ori_df['D, non'] + ori_df['CD, oui'] + ori_df['CD, non'] + ori_df['G, oui'] + ori_df['G, non'])/2
final_ori_df.head(15)
```

```python
final_df = pd.merge(pd.merge(filtered_df, final_ori_df, on='affairVoteId'), full_votes, on='affairVoteId')
final_df.head()
```
